## Exploratory Data Analysis

###### Gender


```python
sns.countplot(x="Gender",data=df)
```

```python
a=sns.countplot(x="Gender",data=df)
for i in a.containers:
    a.bar_label(i)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sale_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum()
sns.barplot(x="Gender",y="Amount",data=sale_gen)
```

### Age

```python
sns.countplot(x="Age Group",hue="Gender",data=df)
```

```python
a=sns.countplot(x="Age Group",hue="Gender",data=df)
for i in a.containers:
    a.bar_label(i)
```

```python
sale_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
age=sns.barplot(x="Age Group",y="Amount",data=sale_age)
```

```python
sale_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(16,5)})
sns.barplot(x="State",y="Orders",data=sale_state)
```

### State

```python
sale_state_amount=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(16,6)})
sns.barplot(x="State",y="Amount",data=sale_state_amount)
```

### Marital Status

```python
ms=sns.countplot(x="Marital_Status",data=df)
sns.set(rc={"figure.figsize":(4,4)})
for i in ms.containers:
    ms.bar_label(i)
```

```python
sale_ms=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(7,4)})
sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=sale_ms)
```

### Occupation

```python
sns.set(rc={"figure.figsize":(20,5)})
o=sns.countplot(x="Occupation",data=df)
for i in o.containers:
    o.bar_label(i)
```

```python
ob=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

obc=sns.barplot(x="Occupation",y="Amount",data=ob)
```

### Product Category

```python
sns.set(rc={"figure.figsize":(25,5)})
obc=sns.countplot(x="Product_Category",data=df)
for i in obc.containers:
    obc.bar_label(i)
```

```python
sns.set(rc={"figure.figsize":(18
                              ,7)})

ob=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(x="Product_Category",y="Amount",data=ob)
```

### Product ID

```python
sns.set(rc={"figure.figsize":(18
                              ,7)})

ob=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(x="Product_ID",y="Orders",data=ob)
```

```python
df.groupby("Product_ID")["Orders"].sum().nlargest(10).plot(kind='bar')
```
