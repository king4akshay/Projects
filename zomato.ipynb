```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(["url","address","phone","menu_item","dish_liked","reviews_list"],axis=1)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["rate"].unique()
```

```python
def hrate(i):
    if(i=="NEW" or i=="-"):
        return np.nan
    else:
        i=str(i).split("/")
        i=i[0]
        return float(i)
df["rate"]=df["rate"].apply(hrate)
df["rate"].head()
```

```python
df.rate.isnull().sum()
```

```python
df["rate"].fillna(df["rate"].mean(),inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.rename(columns={"approx_cost(for two people)":"c2p","listed_in(type)":"type","listed_in(city)":"city"},inplace=True)
```

```python
df
```

```python
df["location"].unique()
```

```python
df["city"].unique()
```

```python
df.drop(["city"],axis=1,inplace=True)
```

```python
df["c2p"].unique()
```

```python
import re
df["c2p"] = df["c2p"].str.replace(",", " ")
df["c2p"] = df["c2p"].replace(r"[^\d.]", "", regex=True)
df["c2p"] = df["c2p"].astype(float)
```

```python
df
```

```python
df["c2p"].unique()
```

## Cleaning rest type

```python
df["rest_type"].unique()
```

```python
rest_types=df["rest_type"].value_counts()
rest_types
```

```python
rest_typeless=rest_types[rest_types<1000]
rest_typeless
```

```python
df["rest_type"].value_counts()
```

```python
def hrest(i):
    if(i in rest_typeless):
        return'others'
    else:
        return i
df["rest_type"]=df["rest_type"].apply(hrest)
df["rest_type"].value_counts()
```

```python
df.head()
```

## Cleaning location

```python
location=df["location"].value_counts()
locationless=location[location<300]
def hloc(i):
    if(i in locationless):
        return'others'
    else:
        return i
df["location"]=df["location"].apply(hloc)
df["location"].value_counts()
```

```python
df.head()
```

## cleaning cuisines

```python
cuisines=df["cuisines"].value_counts()
cuisinesless=cuisines[cuisines<100]
def hcui(i):
    if(i in cuisinesless):
        return'others'
    else:
        return i
df["cuisines"]=df["cuisines"].apply(hcui)
df["cuisines"].value_counts()
```

```python
df.head()
```

```python
df["type"].value_counts()
```
